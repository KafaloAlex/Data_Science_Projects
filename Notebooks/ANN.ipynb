```python
import tensorflow as tf

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
# Chargement du dataset depuis sklearn
iris_data = load_iris()
features = iris_data.data
target = iris_data.target.reshape(-1, 1)
```

```python
target.shape
```

```python
# Encodage de la variable cible
targets = OneHotEncoder().fit_transform(target).toarray()
```

```python
type(targets)
```

```python
# Données de Test et d'entrainement
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2)
```

```python
# Création du perceptron multicouche
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
# Compilation du modèle
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Entrainement du modèle
model.fit(X_train, y_train, epochs=50, batch_size=1, verbose=1)
```

```python
# Evaluation du modèle
taux_erreur, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Taux d'erreur : {taux_erreur}\nAccuracy : {accuracy}")
```
